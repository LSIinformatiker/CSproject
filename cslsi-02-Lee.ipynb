```python
print('write a recursive function for polynomial p(x)')
print('with this formular: a0 + x*(a1+a2*x +a3*x**2 + ... + an*x**(n-1))''\n')
x=input("input number variable x(integer): ")
n=input("input repetation number n(integer): ")
a0=input("please input initial a0 value(integer)(rule: an=a0+n): ")
a0=int(a0)
x=int(x)
n=int(n)
pl=[]
esv=[]
f=0
for i in range(n+1):     ### To solve the list range issue, n+1 
    global pl            ### pl[n+1] not gonna used in the formular
    uaa=a0+i             ### user assigned variable a value
    pl.append(uaa)       ### pl=[a0,a1,a2,a3,...etc],an=n
    ### if user want to give some rules for "an" sequence, modify uaa formular
    ### ex)a0 +-*/ i, i**i, i+5i, i/i+1, etc
print(pl)
def polynominal(x,n,f):
    global pl
    global esv
    if n==0:                  
        px=f+pl[0]      ### first a0=pl[0]
                        ### the last cumulated formular result + a0
                        ### a0 + x*(a1+a2*x +a3*x**2 + ... + an*x**(n-1))
        print('\n'"P(x) the result of recursive function is ")
        return px     
    elif n >= 1:             
        res=(x*((pl[n-1])*(x**(n-1)))) ### to avoid use the pl[n+1] value 
        esv.append(res) ### track the formular result process
        print(esv)       
        f=f+res         ### sum result and hand over to the next value
        print(f)
        return polynominal(x,n-1,f) ### reverse sequence n to the 0(end point)
        ### Recursive function! 
    else:
        print("the n value must be grower than 0, and be the integer")
        n=input("input sequence number n(integer): ")
        return polynominal(x,n,f)
print(polynominal(x,n,f))
```
